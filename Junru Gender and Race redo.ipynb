```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import gender_guesser.detector as gender
gender_retrieve = gender.Detector()
```

```python
all_osha = pd.read_csv('all_OSHA_merged.csv', usecols=['PRINCIPAL','activity_nr', 'open_year','DUNSNO','news_year','5km_ind','press_release_link','estabid_osha','YEAR','current_penalty','latitude','longitude'])
```

```python
all_OSHA_geo = pd.read_csv('OSHA_cases_distant_to_news.csv')
```

```python
all_osha
```

```python
name_to_race = pd.read_csv('names_with_race_junru.csv', usecols=['PRINCIPAL','race'])
name_to_race
```

```python
all_osha['First Name'] = all_osha['PRINCIPAL'].apply(lambda x: x.split()[0] if pd.notnull(x) else x)
all_osha['First Name'] = [i.capitalize() for i in all_osha['First Name']]
all_osha['Last Name'] = all_osha['PRINCIPAL'].apply(lambda x: x.split()[1] if pd.notnull(x) and len(x.split()) > 1 else None)
all_osha['Gender'] = [gender_retrieve.get_gender(i) for i in all_osha['First Name']]
all_osha
```

```python
all_osha = pd.merge(all_osha, name_to_race, how='left', left_on='PRINCIPAL', right_on='PRINCIPAL')
all_osha
```

```python
all_osha_groupby_len = len(all_osha.groupby(['activity_nr', 'DUNSNO']).count())
```

```python
dict_news = {}
counter = 0
for i in range(-7,8):
    dict_news['CEO_change_'+str(i)] = []
    dict_news['penalty_amount_'+str(i)] = []
    dict_news['news_' + str(i)] = []
    dict_news['gender_' + str(i)] = []
    dict_news['race_' + str(i)] = []
for name, group in all_osha.groupby(['activity_nr', 'DUNSNO']):
    #put counter
    try:
        if np.isnan(group['press_release_link'].iloc[0]):
            group['current_penalty'] = group['current_penalty'].fillna(0)
            for i in range(-7,8):
                try:
                    if group[group['YEAR'] == group['open_year'].iloc[0] + i]['PRINCIPAL'].iloc[0] != group[group['YEAR'] == group['open_year'].iloc[0]]['PRINCIPAL'].iloc[0]:
                        dict_news['CEO_change_'+str(i)].append(1)
                    else: 
                        dict_news['CEO_change_'+str(i)].append(0)
                    dict_news['news_'+str(i)].append(0)
                    dict_news['penalty_amount_'+str(i)].append(group['current_penalty'].iloc[0])
                    dict_news['gender_' + str(i)].append(group['Gender'].iloc[0])
                    dict_news['race_' + str(i)].append(group['race'].iloc[0])
                except IndexError:
                    pass
    except TypeError:
        group['current_penalty'] = group['current_penalty'].fillna(0)
        for i in range(-7,8):
            try:
                if group[group['YEAR'] == group['open_year'].iloc[0] + i]['PRINCIPAL'].iloc[0] != group[group['YEAR'] == group['open_year'].iloc[0]]['PRINCIPAL'].iloc[0]:
                    dict_news['CEO_change_'+str(i)].append(1)
                else: 
                    dict_news['CEO_change_'+str(i)].append(0)
                dict_news['news_'+str(i)].append(1)
                dict_news['penalty_amount_'+str(i)].append(group['current_penalty'].iloc[0])
                dict_news['gender_' + str(i)].append(group['Gender'].iloc[0])
                dict_news['race_' + str(i)].append(group['race'].iloc[0])
            except IndexError:
                pass
    counter +=1
    if (counter % 5000) == 0:
        print(str((counter*100)/all_osha_groupby_len) + '% completed')
```

```python
news_data_dict = {}
for i in range(-7,8):
    news_data_dict['news_data_year_'+str(i)] = pd.DataFrame({'news': dict_news['news_'+str(i)], 'change_CEO': dict_news['CEO_change_'+str(i)], 
                                                             'penalty': dict_news['penalty_amount_'+str(i)], 'gender': dict_news['gender_'+str(i)],'race': dict_news['race_'+str(i)]})
```

```python
for i in range(-7,8):
    news_data_dict['news_data_year_'+str(i)].to_csv('CEO_Overturn_News_'+str(i)+'_Year_After_Violation.csv (G+R)')
```

```python
dict_dist_news = {(all_OSHA_geo['latitude'][i], all_OSHA_geo['longitude'][i]): all_OSHA_geo['dist_news'][i] for i in range(len(all_OSHA_geo['dist_news']))}
dict_dist_news_year = {(all_OSHA_geo['latitude'][i], all_OSHA_geo['longitude'][i]): all_OSHA_geo['news_year'][i] for i in range(len(all_OSHA_geo['news_year']))}
```

```python
all_osha['dist_news'] = all_osha.apply(lambda x: dict_dist_news[(x['latitude'], x['longitude'])], axis = 1)
all_osha['news_year'] = all_osha.apply(lambda x: dict_dist_news_year[(x['latitude'], x['longitude'])], axis = 1)
```

```python
all_osha['5km_ind'] = np.where(all_osha['dist_news'] <= 5, 1, 0)
```

```python
dict_dist = {}
counter2 = 0
for i in range(-7,8):
    dict_dist['CEO_change_'+str(i)] = []
    dict_dist['penalty_amount_'+str(i)] = []
    dict_dist['dist_'+str(i)] = []
    dict_dist['dist_ind_'+str(i)] = []
    dict_dist['year_'+str(i)] = []
    dict_dist['act_num_'+str(i)] = []
    dict_dist['OSHA_ID_'+str(i)] = []
    dict_dist['DUNS_'+str(i)] = []
    dict_dist['gender_' + str(i)] = []
    dict_dist['race_' + str(i)] = []
dict_CEO_change = {}
for i in range(-6,8):
    dict_CEO_change['change_in_year_'+str(i)] = []
dict_CEO_change['dist_ind'] = []
dict_CEO_change['act_num'] = []
dict_CEO_change['OSHA_ID'] = []
dict_CEO_change['DUNS'] = []
dict_CEO_change['gender'] = []
dict_CEO_change['race'] = []

for name, group in all_osha.groupby(['activity_nr', 'DUNSNO']):
    #put counter
    dict_CEO_change['OSHA_ID'].append(group['estabid_osha'].iloc[0])
    dict_CEO_change['act_num'].append(name[0])
    dict_CEO_change['DUNS'].append(name[1])
    dict_CEO_change['gender'].append(group['Gender'].iloc[0])
    dict_CEO_change['race'].append(group['race'].iloc[0])
    if group['5km_ind'].iloc[0] == 0 or group['YEAR'].iloc[0]:
        group['current_penalty'] = group['current_penalty'].fillna(0)
        dict_CEO_change['dist_ind'].append(0)
        for i in range(-7,8):
            try:
                if group[group['YEAR'] == group['open_year'].iloc[0] + i]['PRINCIPAL'].iloc[0] != group[group['YEAR'] == group['open_year'].iloc[0]]['PRINCIPAL'].iloc[0]:
                    dict_dist['CEO_change_'+str(i)].append(1)
                else: 
                    dict_dist['CEO_change_'+str(i)].append(0)
                dict_dist['dist_ind_'+str(i)].append(0)
                dict_dist['penalty_amount_'+str(i)].append(group['current_penalty'].iloc[0])
                dict_dist['dist_'+str(i)].append(group['dist_news'].iloc[0])
                dict_dist['year_'+str(i)].append(group['open_year'].iloc[0]+i)
                dict_dist['OSHA_ID_'+str(i)].append(group['estabid_osha'].iloc[0])
                dict_dist['act_num_'+str(i)].append(name[0])
                dict_dist['DUNS_'+str(i)].append(name[1])
                dict_dist['gender_' + str(i)].append(group['Gender'].iloc[0])
                dict_dist['race_' + str(i)].append(group['race'].iloc[0])
            except IndexError:
                pass
            if i>-7:
                try:
                    if group[group['YEAR'] == group['open_year'].iloc[0] + i]['PRINCIPAL'].iloc[0] != group[group['YEAR'] == group['open_year'].iloc[0] + i-1]['PRINCIPAL'].iloc[0]:
                        dict_CEO_change['change_in_year_'+str(i)].append(1)
                    else:
                        dict_CEO_change['change_in_year_'+str(i)].append(0)
                except IndexError:
                    dict_CEO_change['change_in_year_'+str(i)].append(np.nan)

    else:
        group['current_penalty'] = group['current_penalty'].fillna(0)
        dict_CEO_change['dist_ind'].append(1)
        for i in range(-7,8):
            try:
                if group[group['YEAR'] == group['open_year'].iloc[0] + i]['PRINCIPAL'].iloc[0] != group[group['YEAR'] == group['open_year'].iloc[0]]['PRINCIPAL'].iloc[0]:
                    dict_dist['CEO_change_'+str(i)].append(1)
                else: 
                    dict_dist['CEO_change_'+str(i)].append(0)
                dict_dist['dist_ind_'+str(i)].append(1)
                dict_dist['penalty_amount_'+str(i)].append(group['current_penalty'].iloc[0])
                dict_dist['dist_'+str(i)].append(group['dist_news'].iloc[0])
                dict_dist['year_'+str(i)].append(group['YEAR'].iloc[0])
                dict_dist['OSHA_ID_'+str(i)].append(group['estabid_osha'].iloc[0])
                dict_dist['act_num_'+str(i)].append(name[0])
                dict_dist['DUNS_'+str(i)].append(name[1])
                dict_dist['gender_'+str(i)].append(group['Gender'].iloc[0])
                dict_dist['race_'+str(i)].append(group['race'].iloc[0])
            except IndexError:
                pass
            if i>-7:
                try:
                    if group[group['YEAR'] == group['open_year'].iloc[0] + i]['PRINCIPAL'].iloc[0] != group[group['YEAR'] == group['open_year'].iloc[0] + i-1]['PRINCIPAL'].iloc[0]:
                        dict_CEO_change['change_in_year_'+str(i)].append(1)
                    else:
                        dict_CEO_change['change_in_year_'+str(i)].append(0)
                except IndexError:
                    dict_CEO_change['change_in_year_'+str(i)].append(np.nan)
    counter2 +=1
    if (counter2 % 5000) == 0:
        print(str((counter2*100)/all_osha_groupby_len) + '% completed')
```

```python
dist_data_dict = {}
for i in range(-7,8):
    dist_data_dict['dist_data_year_'+str(i)] = pd.DataFrame({'dist_ind': dict_dist['dist_ind_'+str(i)],
                                                             'dist':dict_dist['dist_'+str(i)], 'change_CEO': dict_dist['CEO_change_'+str(i)],
                                                             'penalty': dict_dist['penalty_amount_'+str(i)],
                                                             'YEAR': dict_dist['year_'+str(i)], 'activity_nr': dict_dist['act_num_'+str(i)], 
                                                             'DUNSNO': dict_dist['DUNS_'+str(i)], 'estabid_osha': dict_dist['OSHA_ID_'+str(i)],
                                                            'gender': dict_dist['gender_' + str(i)],'race':dict_dist['race_' + str(i)]})
```

```python
for i in range(-7,8):
    dist_data_dict['dist_data_year_'+str(i)].to_csv('CEO_Overturn_Processed_'+str(i)+'_Year_After_Violation.csv (G+R)', index = False)
```
